```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#1.	What is the major reason people being kidnapped in each state

crime = pd.read_csv('C:\\Users\\Administrator\\Documents\\Data Visualization\\Data Set\\rajanand-crime-in-india\\39_Specific_purpose_of_kidnapping_and_abduction.csv')

crime = crime[0:3230][['Area_Name', 'Year', 'Group_Name', 'K_A_Grand_Total']]

crime.columns = (['State', 'Year', 'Reason', 'Total'])

crime['Reason'] = crime.Reason.replace('Kidnap - For', '', regex = True)

crime.dropna()
k = crime.groupby(['State', 'Reason'])['Total'].sum().reset_index()

df = k.pivot('State', 'Reason', 'Total')
sb.heatmap(df, linewidths = 0.3, center = 1)
```
