```python
terror = pd.read_csv('C:\\Users\\Administrator\\Desktop\\Python Assignments\\DataSets\\terror.csv', encoding='latin', low_memory=False)
terror.head()
```

```python
#Q.1)	How many attacks happened in India?
I = terror[terror.country_txt == 'India']
att = I.groupby('country_txt')['country'].value_counts()
print(att)
```

```python
#Q.2)	How many attacks happened in India and upto 3 people were killed? 
t = terror[terror.country_txt == 'India'][terror.nkill == 3]
t.groupby('country_txt')['nkill'].value_counts()
```

```python
#Q.3)	Extract the city and summary for attacks above
print('The cities in which attacks happened in India and upto 3 people were killed are\n\n%s'% (t.city))
print('\n\nSummary for above attacks is\n\n',t.describe())
```

```python
#Q.4)	In a single terror incident in India, find out top 5 cities by number killed
top = I.groupby('city')['nkill'].max()
t5 = top.sort_values(ascending = False)
t5.head()
```

```python
#Q.5)	In a single terror incident in India, find out top 5 cities by number killed and wounded 
topp = I.groupby('city')[['nkill', 'nwound']].max()
kw = topp.sort_values('nwound', ascending = False)
kw.head()
```
